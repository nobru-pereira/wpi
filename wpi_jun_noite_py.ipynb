```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #biblioteca para apresentar gr√°fico
```

```python
### dataset #a212: amostra..#

a212_b = pd.read_csv('/home/bruno/git/wpi/10junho2021noite/aB212.CSV' ,skiprows=15)
a53_b = pd.read_csv('/home/bruno/git/wpi/10junho2021noite/aB53.CSV' ,skiprows=15)
a83_b = pd.read_csv('/home/bruno/git/wpi/10junho2021noite/aB83.CSV' ,skiprows=15)
a207_b = pd.read_csv('/home/bruno/git/wpi/10junho2021noite/aB207.CSV' ,skiprows=15)
a67_b = pd.read_csv('/home/bruno/git/wpi/10junho2021noite/aB67.CSV' ,skiprows=15)
asar_b = pd.read_csv('/home/bruno/git/wpi/10junho2021noite/asar_b.CSV' ,skiprows=15)
asdk_b = pd.read_csv('/home/bruno/git/wpi/10junho2021noite/asardrk_b.CSV' ,skiprows=15)
asaref_b = pd.read_csv('/home/bruno/git/wpi/10junho2021noite/asarref_b.CSV' ,skiprows=11)
dk_b = pd.read_csv('/home/bruno/git/wpi/10junho2021noite/drk_b.CSV' ,skiprows=11)
ref_b = pd.read_csv('/home/bruno/git/wpi/10junho2021noite/ref_b.CSV' ,skiprows=11)
```

```python
### index wavelenght column

df_a212_b = a212_b.set_index('wavelenght')
df_a53_b = a53_b.set_index('wavelenght')
df_a83_b = a83_b.set_index('wavelenght')
df_a207_b = a207_b.set_index('wavelenght')
df_a67_b = a67_b.set_index('wavelenght')
df_asar_b = asar_b.set_index('wavelenght')
df_asaref_b = asaref_b.set_index('wavelenght')
df_dk_b = dk_b.set_index('wavelenght')
df_ref_b = ref_b.set_index('wavelenght')
```

```python
### stats

df_a212_b['mean'] = df_a212_b.mean(axis=1)
df_a53_b['mean'] = df_a53_b.mean(axis=1)
df_a83_b['mean'] = df_a83_b.mean(axis=1)
#df_dk['mean'] = df_dk.mean(axis=1)
df_a207_b['mean'] = df_a207_b.mean(axis=1)
df_a67_b['mean'] = df_a67_b.mean(axis=1)
df_asar_b['mean'] = df_asar_b.mean(axis=1)

df_a212_b['median'] = df_a212_b.median(axis=1)
df_a53_b['median'] = df_a53_b.median(axis=1)
df_a83_b['median'] = df_a83_b.median(axis=1)
#df_dk['mean'] = df_dk.mean(axis=1)
df_a207_b['median'] = df_a207_b.median(axis=1)
df_a67_b['median'] = df_a67_b.median(axis=1)
df_asar_b['median'] = df_asar_b.median(axis=1)



df_a212_b['std'] = df_a212_b.std(axis=1)
df_a53_b['std'] = df_a53_b.std(axis=1)
df_a83_b['std'] = df_a83_b.std(axis=1)
#df_dk['mean'] = df_dk.mean(axis=1)
df_a207_b['std'] = df_a207_b.std(axis=1)
df_a67_b['std'] = df_a67_b.std(axis=1)
df_asar_b['std'] = df_asar_b.std(axis=1)
```

```python
fig, ax = plt.subplots(figsize=(8,4))
# plt.plot(df_ta['Wavelenght'], df1['E1'], label='E1')
# plt.plot(df1['Wavelenght'], df1['E2'], label='E2')
# plt.plot(df1['Wavelenght'], df1['E3'], label='E3')
# plt.plot(df1['Wavelenght'], df1['E4'], label='E4')
# plt.plot(df1['Wavelenght'], df1['E5'], label='E5')
plt.plot(df_a53_b['mean'], ':', label='Mean')
plt.plot(df_a53_b['median'], '--', label='Median', color='black')
plt.plot(df_a53_b['std'], ':', label='Standard Deviation', color='gray')
plt.title('10 jun noite - a53_b - data')
plt.grid(linestyle='--')
plt.legend()
```
