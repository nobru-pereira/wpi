```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #biblioteca para apresentar gr√°fico
```

```python
### dataset #ta: teste a, tb..#

ta = pd.read_csv('/home/bruno/git/wpi/10junho2021/testea1.CSV' ,skiprows=15)
tb = pd.read_csv('/home/bruno/git/wpi/10junho2021/testeb1.CSV' ,skiprows=15)
tc = pd.read_csv('/home/bruno/git/wpi/10junho2021/testec1.CSV' ,skiprows=15)
dk = pd.read_csv('/home/bruno/git/wpi/10junho2021/drk1.CSV' ,skiprows=11)
mq = pd.read_csv('/home/bruno/git/wpi/10junho2021/miliq1_1.CSV' ,skiprows=15)
mq_a = pd.read_csv('/home/bruno/git/wpi/10junho2021/miliqaftertestea1.CSV' ,skiprows=15)
mq_b = pd.read_csv('/home/bruno/git/wpi/10junho2021/miliqaftertesteb1.CSV' ,skiprows=15)
mq_c = pd.read_csv('/home/bruno/git/wpi/10junho2021/miliqaftertestec1.CSV' ,skiprows=15)
```

```python
### index wavelenght column

df_ta = ta.set_index('wavelenght')
df_tb = tb.set_index('wavelenght')
df_tc = tc.set_index('wavelenght')
df_dk = dk.set_index('wavelenght')
df_mq = mq.set_index('wavelenght')
df_mqa = mq_a.set_index('wavelenght')
df_mqb = mq_b.set_index('wavelenght')
df_mqc = mq_c.set_index('wavelenght')

### stats

df_ta['mean'] = df_ta.mean(axis=1)
df_tb['mean'] = df_tb.mean(axis=1)
df_tc['mean'] = df_tc.mean(axis=1)
#df_dk['mean'] = df_dk.mean(axis=1)
df_mq['mean'] = df_mq.mean(axis=1)
df_mqa['mean'] = df_mqa.mean(axis=1)
df_mqb['mean'] = df_mqb.mean(axis=1)
df_mqc['mean'] = df_mqc.mean(axis=1)

df_ta['median'] = df_ta.median(axis=1)
df_tb['median'] = df_tb.median(axis=1)
df_tc['median'] = df_tc.median(axis=1)
#df_dk['median'] = df_dk.median(axis=1)
df_mq['median'] = df_mq.median(axis=1)
df_mqa['median'] = df_mqa.median(axis=1)
df_mqb['median'] = df_mqb.median(axis=1)
df_mqc['median'] = df_mqc.median(axis=1)

df_ta['std'] = df_ta.std(axis=1)
df_tb['std'] = df_tb.std(axis=1)
df_tc['std'] = df_tc.std(axis=1)
#df_dk['std'] = df_dk.std(axis=1)
df_mq['std'] = df_mq.std(axis=1)
df_mqa['std'] = df_mqa.std(axis=1)
df_mqb['std'] = df_mqb.std(axis=1)
df_mqc['std'] = df_mqc.std(axis=1)
```

```python
fig, ax = plt.subplots(figsize=(8,4))
# plt.plot(df_ta['Wavelenght'], df1['E1'], label='E1')
# plt.plot(df1['Wavelenght'], df1['E2'], label='E2')
# plt.plot(df1['Wavelenght'], df1['E3'], label='E3')
# plt.plot(df1['Wavelenght'], df1['E4'], label='E4')
# plt.plot(df1['Wavelenght'], df1['E5'], label='E5')
plt.plot(df_ta['mean'], ':', label='Mean')
plt.plot(df_ta['median'], '--', label='Median', color='black')
plt.plot(df_ta['std'], ':', label='Standard Deviation', color='gray')
plt.title('10 jun manha - a - data')
plt.grid(linestyle='--')
plt.legend()
```
