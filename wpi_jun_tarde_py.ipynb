```python
### dataset #a212: amostra..#

a212 = pd.read_csv('/home/bruno/git/wpi/10junho2021tarde/amostr212_1.CSV' ,skiprows=15)
a53 = pd.read_csv('/home/bruno/git/wpi/10junho2021tarde/amostra53_1.CSV' ,skiprows=15)
a83 = pd.read_csv('/home/bruno/git/wpi/10junho2021tarde/amostra83_1.CSV' ,skiprows=15)
a207 = pd.read_csv('/home/bruno/git/wpi/10junho2021tarde/amostra207_1.CSV' ,skiprows=15)
a67 = pd.read_csv('/home/bruno/git/wpi/10junho2021tarde/amostra67_1.CSV' ,skiprows=15)
asar = pd.read_csv('/home/bruno/git/wpi/10junho2021tarde/asar1.CSV' ,skiprows=15)
asdk = pd.read_csv('/home/bruno/git/wpi/10junho2021tarde/asardrk1.CSV' ,skiprows=15)
asaref = pd.read_csv('/home/bruno/git/wpi/10junho2021tarde/asarref1.CSV' ,skiprows=11)
dk = pd.read_csv('/home/bruno/git/wpi/10junho2021tarde/drk1.CSV' ,skiprows=11)
ref = pd.read_csv('/home/bruno/git/wpi/10junho2021tarde/ref1.CSV' ,skiprows=11)
```

```python
### index wavelenght column

df_a212 = a212.set_index('wavelenght')
df_a53 = a53.set_index('wavelenght')
df_a83 = a83.set_index('wavelenght')
df_a207 = a207.set_index('wavelenght')
df_a67 = a67.set_index('wavelenght')
df_asar = asar.set_index('wavelenght')
df_asaref = asaref.set_index('wavelenght')
df_dk = dk.set_index('wavelenght')
df_ref = ref.set_index('wavelenght')
```

```python
### stats

df_a212['mean'] = df_a212.mean(axis=1)
df_a53['mean'] = df_a53.mean(axis=1)
df_a83['mean'] = df_a83.mean(axis=1)
#df_dk['mean'] = df_dk.mean(axis=1)
df_a207['mean'] = df_a207.mean(axis=1)
df_a67['mean'] = df_a67.mean(axis=1)
df_asar['mean'] = df_asar.mean(axis=1)

df_a212['median'] = df_a212.median(axis=1)
df_a53['median'] = df_a53.median(axis=1)
df_a83['median'] = df_a83.median(axis=1)
#df_dk['mean'] = df_dk.mean(axis=1)
df_a207['median'] = df_a207.median(axis=1)
df_a67['median'] = df_a67.median(axis=1)
df_asar['median'] = df_asar.median(axis=1)



df_a212['std'] = df_a212.std(axis=1)
df_a53['std'] = df_a53.std(axis=1)
df_a83['std'] = df_a83.std(axis=1)
#df_dk['mean'] = df_dk.mean(axis=1)
df_a207['std'] = df_a207.std(axis=1)
df_a67['std'] = df_a67.std(axis=1)
df_asar['std'] = df_asar.std(axis=1)
```

```python
fig, ax = plt.subplots(figsize=(8,4))
# plt.plot(df_ta['Wavelenght'], df1['E1'], label='E1')
# plt.plot(df1['Wavelenght'], df1['E2'], label='E2')
# plt.plot(df1['Wavelenght'], df1['E3'], label='E3')
# plt.plot(df1['Wavelenght'], df1['E4'], label='E4')
# plt.plot(df1['Wavelenght'], df1['E5'], label='E5')
plt.plot(df_a53['mean'], ':', label='Mean')
plt.plot(df_a53['median'], '--', label='Median', color='black')
plt.plot(df_a53['std'], ':', label='Standard Deviation', color='gray')
plt.title('10 jun tarde - a53 - data')
plt.grid(linestyle='--')
plt.legend()
```
